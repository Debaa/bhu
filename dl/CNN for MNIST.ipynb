```python
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
# from keras import backend as K
```

# Splitting the MNIST dataset into Train and Test

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
import numpy as np
np.set_printoptions(linewidth=1000)  # increase width

print(x_train.shape)
print(x_train[0].shape)

print(y_train[0])
print(x_train[0])
```

# Preprocessing the input data

```python
num_of_trainImgs = x_train.shape[0] #60000 here
num_of_testImgs = x_test.shape[0] #10000 here
img_width = 28
img_height = 28
 
x_train = x_train.reshape(x_train.shape[0], img_height, img_width, 1)
x_test = x_test.reshape(x_test.shape[0], img_height, img_width, 1)
print(x_train[0])
input_shape = (img_height, img_width, 1)
 
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
```

```python


print(input_shape)
```

```python
import numpy as np
np.set_printoptions(linewidth=1000)  # increase width

print(x_train.shape)
print(x_train[0].shape)

print(y_train[10000])
print(x_train[10000])
```

# Converting the class vectors to binary class

```python
num_classes = 10
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

# Defining the model architecture

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=input_shape))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))
```

# Compiling the model

```python
model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])
```

# Fitting the model on training data

```python
model.fit(x_train, y_train,
          batch_size=228,
          epochs=12,
          verbose=1,
          validation_data=(x_test, y_test))
```

#  Evaluating the model on test data

```python
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

# Saving the Model

```python
model.save('trained_model.h5')
```
