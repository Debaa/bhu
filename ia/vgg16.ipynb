```python
# ==========================================
# VGG16 - Bird Classification (10 Classes)
# ==========================================
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import load_img, img_to_array
# ------------------------------------------
# Dataset Folder Structure
# ------------------------------------------
# data/
#    bird1/
#    bird2/
#    bird3/
#    ...
#    bird10/

# ------------------------------------------
# Load Dataset
# ------------------------------------------

train_data = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train = train_data.flow_from_directory(
    "data",
    target_size=(224,224),
    batch_size=32,
    subset="training"
)

test = train_data.flow_from_directory(
    "data",
    target_size=(224,224),
    batch_size=32,
    subset="validation"
)

# ------------------------------------------
# Load VGG16
# ------------------------------------------

base_model = VGG16(
    weights='imagenet',
    include_top=False,
    input_shape=(224,224,3)
)

base_model.trainable = False

# ------------------------------------------
# Build Model
# ------------------------------------------

model = Sequential([
    base_model,
    Flatten(),
    Dense(10, activation='softmax')
])

# ------------------------------------------
# Compile Model
# ------------------------------------------

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# ------------------------------------------
# Train Model
# ------------------------------------------

#model.fit(train, validation_data=test, epochs=3)

# ------------------------------------------
# Save Model
# ------------------------------------------


#model.save("bird_model.h5")
```
